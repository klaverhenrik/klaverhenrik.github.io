```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
print(cv2.__version__)
print(tf.__version__)
```

# Load images

```python
from os import listdir
from os.path import isfile, join
img_files = [f for f in listdir('images') if isfile(join('images', f))]
no_frames = len(img_files)
images = []
for frame in range(0,no_frames):
    filename = 'images/frame' + str(frame+1) + '.png'
    img = cv2.imread(filename)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = img.astype(np.float32) / 255.
    images.append(img)
```

# Reference image

```python
im1 = images[0]
im1_filt = cv2.GaussianBlur(im1,(3,3),0)
Ix = np.gradient(im1_filt, axis=1)
Iy = np.gradient(im1_filt, axis=0)
plt.figure()
plt.subplot(221)
plt.imshow(im1, cmap='gray')
plt.subplot(222)
plt.imshow(im1_filt, cmap='gray')
plt.subplot(223)
plt.imshow(Ix, cmap='gray')
plt.subplot(224)
plt.imshow(Iy, cmap='gray')
```

# Binary mask

```python
xmin = 24
ymin = 14
width = 14
height = 12

mask = np.zeros(im1.shape, dtype=np.float32)
mask[ymin:ymin+height,xmin:xmin+width] = 1.

mask_flat = mask.flatten()
sel_ix = np.where(mask_flat > 0)
num_pix = sel_ix[0].shape[0]
#mask_flat_reshape = mask_flat.reshape(mask.shape)

plt.figure()
plt.imshow(np.multiply(im1,mask), cmap='gray')
```

# Grid

```python
N = mask.shape[0]
M = mask.shape[1]
x = np.arange(M)
y = np.arange(N)
X, Y = np.meshgrid(x, y)
X = X.astype(np.float32)
Y = Y.astype(np.float32)
```

```python
Ix_flat = Ix.flatten()
Iy_flat = Iy.flatten()
X_flat = X.flatten()
Y_flat = Y.flatten()
A = np.zeros((num_pix,6), dtype=np.float32)
b = np.zeros((num_pix,1), dtype=np.float32)
A[:,0] = Ix_flat[sel_ix]
A[:,1] = np.multiply(Ix_flat[sel_ix],X_flat[sel_ix])
A[:,2] = np.multiply(Ix_flat[sel_ix],Y_flat[sel_ix])
A[:,3] = Iy_flat[sel_ix]
A[:,4] = np.multiply(Iy_flat[sel_ix],X_flat[sel_ix])
A[:,5] = np.multiply(Iy_flat[sel_ix],Y_flat[sel_ix])
```

```python
no_frames = len(images);
res = []
orig = []
flowX = []
flowY = []
res.append(images[0])
orig.append(images[0])
flowX.append(0)
flowY.append(0)
U0 = np.zeros(mask.shape, dtype=np.float32)
V0 = np.zeros(mask.shape, dtype=np.float32)
for frame in range(1, no_frames):
    im2 = images[frame]

    # apply accumulated warp
    X2 = X + U0;
    Y2 = Y + V0;
    im2_warped = cv2.remap(im2, X2, Y2, cv2.INTER_LINEAR)
    
    # filter image and calculate It
    im2_filt = cv2.GaussianBlur(im2_warped,(3,3),0)
    It = im2_filt-im1_filt;
    It_flat = It.flatten()
    
    # Calculate OF
    b = -It_flat[sel_ix]
    a = np.dot(np.dot(np.linalg.pinv(np.dot(A.transpose(),A)),A.transpose()),b)
    
    #U = np.zeros(mask.shape, dtype=np.float32)
    #V = np.zeros(mask.shape, dtype=np.float32)
    U = a[0] + a[1]*X + a[2]*Y;
    V = a[3] + a[4]*X + a[5]*Y;
    
    # Warp im1 to fit with im2
    X2 = X + U0 + U
    Y2 = Y + V0 + V
    im2_warped = cv2.remap(im2, X2, Y2, cv2.INTER_LINEAR)
    
    res.append(im2_warped)
    orig.append(im2)
    
    U0 = U0 + U
    V0 = V0 + V
    
    flowX.append(U0)
    flowY.append(V0)
```

```python
for frame in range(0, no_frames):
    plt.figure()
    plt.subplot(141); plt.imshow(orig[frame], cmap='gray')
    plt.subplot(142); plt.imshow(res[frame], cmap='gray')
    plt.subplot(143); plt.imshow(np.multiply(mask,orig[frame]), cmap='gray')
    plt.subplot(144); plt.imshow(np.multiply(mask,res[frame]), cmap='gray')
```

```python
tf.reset_default_graph()
sess = tf.InteractiveSession()
```

```python
def reg_error():
    img = tf.placeholder(tf.float32,shape=((1,N,M,1)),name='img')
    ref = tf.placeholder(tf.float32,shape=((1,N,M,1)),name='ref')
    flow = tf.placeholder(tf.float32,shape=((1,N,M,2)),name='flow')
    img_warped = tf.contrib.image.dense_image_warp(img,flow)
    loss = tf.losses.mean_squared_error(img_warped,ref)
    return { 'loss': loss,
        'img': img,
        'ref': ref,
        'flow': flow,
        'img_warped': img_warped}

loss = reg_error()

for frame in range(1, no_frames):
    img_np = np.reshape(images[frame],(1,N,M,1)).astype(np.float32)
    ref_np = np.reshape(images[0],(1,N,M,1)).astype(np.float32)
    flow_np = np.zeros((1,N,M,2), dtype=np.float32)
    flow_np[0,:,:,0] = -flowY[frame]
    flow_np[0,:,:,1] = -flowX[frame]
    feed_dict = {
        loss['img']: img_np,
        loss['ref']: ref_np,
        loss['flow']: flow_np
    }
    im_warp, loss_val = sess.run([loss['img_warped'],loss['loss']], feed_dict=feed_dict)
    im_warp = im_warp.squeeze()
    plt.figure()
    plt.subplot(141)
    plt.imshow(im_warp, cmap='gray')
    plt.subplot(142)
    plt.imshow(res[frame], cmap='gray')
    plt.subplot(143)
    plt.imshow(np.multiply(mask,im_warp), cmap='gray')
    plt.subplot(144)
    plt.imshow(np.multiply(mask,res[frame]), cmap='gray')
```

```python
learning_rate = 0.01
global_step_encoder = tf.Variable(0, trainable=False)
starter_learning_rate = learning_rate
learning_rate_encoder = tf.train.exponential_decay(starter_learning_rate, global_step_encoder, 100, 0.9, staircase=True)

scale_factor = 4
img_highres = tf.Variable(np.random.uniform(size=(1,scale_factor*N,scale_factor*M,1)), name = "img_highres")
img_highres_tile = tf.tile(img_highres,[no_frames,1,1,1])
ref = tf.placeholder(tf.float32,shape=((no_frames,N,M,1)),name='img')
mask_tile = tf.placeholder(tf.float32,shape=((no_frames,N,M,1)),name='mask')
flow = tf.placeholder(tf.float32,shape=((no_frames,N,M,2)),name='flow')
flow_highres = tf.image.resize_bilinear(
    scale_factor*flow,
    [scale_factor*N,scale_factor*M]
)
img_warped = tf.contrib.image.dense_image_warp(img_highres_tile,flow_highres)
img_warped = tf.image.resize_bilinear(
    img_warped,
    [N,M]
)
cost = tf.losses.mean_squared_error(tf.multiply(img_warped,mask_tile),tf.multiply(ref,mask_tile))

optimizer_encoder = tf.train.AdamOptimizer(learning_rate = learning_rate)
train_op_encoder = optimizer_encoder.minimize(loss=cost,global_step=global_step_encoder)

sess.run(tf.global_variables_initializer())

ref_np = np.zeros((no_frames,N,M,1), dtype=np.float32)
flow_np = np.zeros((no_frames,N,M,2), dtype=np.float32)
mask_np = np.zeros((no_frames,N,M,1), dtype=np.float32)
ref_np[0,:,:,0] = images[0]
flow_np[0,:,:,:] = 0
mask_np[0,:,:,0] = mask
for frame in range(1, no_frames):    
    ref_np[frame,:,:,0] = images[frame]
    flow_np[frame,:,:,0] = flowY[frame]
    flow_np[frame,:,:,1] = flowX[frame]
    mask_np[frame,:,:,0] = mask

feed_dict = {
        ref: ref_np,
        flow: flow_np,
        mask_tile: mask_np
}
loss_val = sess.run(cost, feed_dict=feed_dict)
```

```python
img_warped_np = sess.run(img_warped, feed_dict=feed_dict)
img_warped_np = np.squeeze(img_warped_np)
for frame in range(0, no_frames):
    plt.figure()
    plt.subplot(141)
    plt.imshow(np.squeeze(img_warped_np[frame,:,:]), cmap='gray')
    plt.subplot(142)
    plt.imshow(images[frame], cmap='gray')
    plt.subplot(143)
    plt.imshow(np.multiply(mask,np.squeeze(img_warped_np[frame,:,:])), cmap='gray')
    plt.subplot(144)
    plt.imshow(np.multiply(mask,images[frame]), cmap='gray')
```

```python
for iter in range(1000):
    lr,_,loss_val = sess.run([learning_rate_encoder,train_op_encoder,cost], feed_dict=feed_dict)
    gs = tf.train.global_step(sess, global_step_encoder)
    if iter%100 == 0:
        out = sess.run(img_highres)
        out = out.squeeze()
        plt.figure()
        plt.imshow(out, cmap='gray')
        print("Global step %i, %f, %s=%e" % (gs, lr, 'loss',loss_val))
        plt.show()
```

```python
filename = 'examples.png'
img = cv2.imread(filename)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = img.astype(np.float32) / 255.
img = img[80:200,400:600]
im1 = cv2.resize(img,None,fx=0.25, fy=0.25)
print(img.shape)
print(im1.shape)
plt.figure()
plt.imshow(img,cmap='gray')
plt.figure()
plt.imshow(im1,cmap='gray')
```

```python
N = img.shape[0]
M = img.shape[1]
x = np.arange(M)
y = np.arange(N)
X, Y = np.meshgrid(x, y)
X = X.astype(np.float32)
Y = Y.astype(np.float32)
```

```python
no_frames = 20;
images = []
flowX = []
flowY = []
images.append(im1)
U0 = np.zeros(img.shape, dtype=np.float32)
V0 = np.zeros(img.shape, dtype=np.float32)
flowX.append(U0)
flowY.append(V0)

for frame in range(1, no_frames):
    
    U = np.ones(img.shape, dtype=np.float32) * np.random.rand(1).astype(np.float32) * 8
    V = np.ones(img.shape, dtype=np.float32) * np.random.rand(1).astype(np.float32) * 8
    
    # apply accumulated warp
    X2 = X - U;
    Y2 = Y - V;
    im2 = cv2.remap(img, X2, Y2, cv2.INTER_LINEAR)
    im2 = cv2.resize(im2,None,fx=0.25, fy=0.25)
    
    images.append(im2)
    
    flowX.append(U)
    flowY.append(V)
    
    plt.figure()
    plt.subplot(121); plt.imshow(im1, cmap='gray')
    plt.subplot(122); plt.imshow(images[frame], cmap='gray')
```

```python
tf.reset_default_graph()
sess = tf.InteractiveSession()
```

```python
N2 = im1.shape[0]
M2 = im1.shape[1]
def reg_error():
    img = tf.placeholder(tf.float32,shape=((1,N2,M2,1)),name='img')
    ref = tf.placeholder(tf.float32,shape=((1,N2,M2,1)),name='ref')
    flow = tf.placeholder(tf.float32,shape=((1,N2,M2,2)),name='flow')
    img_warped = tf.contrib.image.dense_image_warp(img,flow)
    loss = tf.losses.mean_squared_error(img_warped,ref)
    return { 'loss': loss,
        'img': img,
        'ref': ref,
        'flow': flow,
        'img_warped': img_warped}

loss = reg_error()

for frame in range(1, no_frames):
    img_np = np.reshape(images[frame],(1,N2,M2,1)).astype(np.float32)
    ref_np = np.reshape(images[0],(1,N2,M2,1)).astype(np.float32)
    flow_np = np.zeros((1,N2,M2,2), dtype=np.float32)
    flow_np[0,:,:,0] = -cv2.resize(0.25*flowY[frame],None,fx=0.25, fy=0.25)
    flow_np[0,:,:,1] = -cv2.resize(0.25*flowX[frame],None,fx=0.25, fy=0.25)
    feed_dict = {
        loss['img']: img_np,
        loss['ref']: ref_np,
        loss['flow']: flow_np
    }
    im_warp, loss_val = sess.run([loss['img_warped'],loss['loss']], feed_dict=feed_dict)
    im_warp = im_warp.squeeze()
    plt.figure()
    plt.subplot(121)
    plt.imshow(im_warp, cmap='gray')
    plt.subplot(122)
    plt.imshow(images[frame], cmap='gray')
```

```python
learning_rate = 0.01
global_step_encoder = tf.Variable(0, trainable=False)
starter_learning_rate = learning_rate
learning_rate_encoder = tf.train.exponential_decay(starter_learning_rate, global_step_encoder, 100, 0.9, staircase=True)

img_highres = tf.Variable(np.random.uniform(size=(1,N,M,1)), name = "img_highres")
img_highres_tile = tf.tile(img_highres,[no_frames,1,1,1])
ref = tf.placeholder(tf.float32,shape=((no_frames,N2,M2,1)),name='img')
flow = tf.placeholder(tf.float32,shape=((no_frames,N,M,2)),name='flow')
img_warped = tf.contrib.image.dense_image_warp(img_highres_tile,flow)
img_warped = tf.image.resize_bilinear(
    img_warped,
    [N2,M2]
)
cost = tf.losses.mean_squared_error(img_warped,ref)

optimizer_encoder = tf.train.AdamOptimizer(learning_rate = learning_rate)
train_op_encoder = optimizer_encoder.minimize(loss=cost,global_step=global_step_encoder)

sess.run(tf.global_variables_initializer())

ref_np = np.zeros((no_frames,N2,M2,1), dtype=np.float32)
flow_np = np.zeros((no_frames,N,M,2), dtype=np.float32)
ref_np[0,:,:,0] = images[0]
flow_np[0,:,:,:] = 0
for frame in range(1, no_frames):    
    ref_np[frame,:,:,0] = images[frame]
    flow_np[frame,:,:,0] = flowY[frame]
    flow_np[frame,:,:,1] = flowX[frame]

feed_dict = {
        ref: ref_np,
        flow: flow_np
}
loss_val = sess.run(cost, feed_dict=feed_dict)
```

```python
for iter in range(1000):
    lr,_,loss_val = sess.run([learning_rate_encoder,train_op_encoder,cost], feed_dict=feed_dict)
    gs = tf.train.global_step(sess, global_step_encoder)
    if iter%100 == 0:
        out = sess.run(img_highres)
        out = out.squeeze()
        plt.figure()
        plt.imshow(out, cmap='gray')
        print("Global step %i, %f, %s=%e" % (gs, lr, 'loss',loss_val))
        plt.show()
```
